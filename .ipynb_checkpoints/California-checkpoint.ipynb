# Project: DATA VISUALIZATION - CALIFORNIA CITIES


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("california_cities.csv")
cities
```

```python
plt.style.use('seaborn')
```

```python
lat ,long = cities['latd'],cities['longd']
population, area =cities['population_total'],cities['area_total_km2']

plt.figure(figsize =(10,6))
plt.scatter(long,lat,c=np.log10(population),cmap='viridis',alpha=0.5,s=area);


plt.title('California Cities: Population and Area Distribution')
plt.xlabel("Latitude");
plt.ylabel("Longtitude");
plt.colorbar(label = 'log$_{10}$(population)');
plt.clim(3,7)

area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,label =str(area)+('km$^2$'),c="royalblue");
plt.legend(labelspacing = 2, title='City Area');
```
